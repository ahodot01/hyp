```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

ins = pd.read_csv("insurance.csv")

# used this line below to see columns and figure out their position
# print(ins.iloc[::]) 

# OUTPUT: SCATTERPLOT FROM INSURANCE.CSV

x = ins.iloc[:,:1].values # values from first column (age), ignoring index 
y = ins.iloc[:,-1].values # values from last dolumn (charges)

plt.scatter(x,y,color = 'c') # green
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

# OUTPUT: SCATTERPLOT WITH BEST-FIT

charges = LinearRegression() # fitting regression model
charges.fit(x,y)

y_predict = charges.predict(x) 

plt.scatter(x,y,color = 'b') # b for blue
plt.plot(x,y_predict,color = 'y') # y for yellow
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

# OUTPUT: SCATTERPLOT WITH BEST-FIT AND PREDICTION FOR UNKNOWN VALUE

unk_charges = [[70]] # unknown value of age

x_predict = np.append(x, unk_charges).reshape(-1,1) 
y_predict = charges.predict(x_predict) 

plt.scatter(x,y,color = 'r') # r for red
plt.plot(x_predict,y_predict,color = 'b') # b for blue
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print(f"Insurance charges at the age of 70 should be: {charges.predict(unk_charges)}")
```
